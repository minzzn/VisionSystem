```python
path = '/content/gdrive/MyDrive/Image_Processing/512test1star.bmp'
# 그레이 스케일로 읽어들이기
image = cv2.imread(path, cv2.IMREAD_GRAYSCALE)

th, blacknwhiteimg = cv2.threshold(image, 0, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)
# 임계값을 받는 th
print('otsu: ', th)
# 131 ~ 255 배경으로 찾고 나머지를 오브젝트로 잡고있다.

cv2_imshow(image)
cv2_imshow(blacknwhiteimg)
print(image.shape)
print(image.size)
print(np.unique(image))
px = image
print(px)
```
