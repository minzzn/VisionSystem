```python
import cv2
import numpy as np
# 과제 2
## 1. RGB 이미지를 읽은 후 화소값 출력

path = '/content/gdrive/MyDrive/차량비전시스템/VisionSystem/Assignment2_Image/girl.jpg'
image=cv2.imread(path,cv2.IMREAD_COLOR)
# print("RGB 이미지의 화소값:",np.unique(image))

# ① Red, Green, Blue 각 채널별 화소값 출력할 것
#  제출물 - 읽어온 RGB 이미지, 채널별 이미지, 채널별 화소값
print("① Red, Green, Blue 각 채널별 화소값 출력할 것")

print(" 제출물")

print(" 읽어온 RGB 이미지 :")
cv2_imshow(image)

print(" 채널별 이미지 :")
b, g, r = cv2.split(image)
cv2_imshow(b)
cv2_imshow(g)
cv2_imshow(r)

print(" 채널별 화소값 :")
print("Red 채널 화소값:", np.unique(r))
print("Green 채널 화소값:", np.unique(g))
print("Blue 채널 화소값:", np.unique(b))
```

```python
# ② HSV 모델로 이미지 표현 및 각 채널별 화소값 출력
#  제출물 - HSV 모델로 변환한 이미지 및 H, S, V 채널별 각 화소값

print("② HSV 모델로 이미지 표현 및 각 채널별 화소값 출력")

print(" 제출물")

print(" HSV 모델로 변환한 이미지 :")
image_hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
cv2_imshow(image_hsv)

print(" H, S, V 채널별 각 화소값 :")
h, s, v = cv2.split(image_hsv)
print("Hue 채널 화소값:", np.unique(h))
print("Saturation 채널 화소값:", np.unique(s))
print("Value 채널 화소값:", np.unique(v))
```

```python
# ③ Gray scale로 변환 후 화소값 출력할 것
#  제출물 - Gray scale로 변환한 이미지 및 해당 화소값

print("③ Gray scale로 변환 후 화소값 출력할 것")

print(" 제출물")

print(" Gray scale로 변환한 이미지 :")
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_imshow(image_gray)

print(" Gray scale 이미지의 화소값:", np.unique(image_gray))
```

```python
## 2. Binary 이미지 두개를 선택
path1 = '/content/gdrive/MyDrive/차량비전시스템/VisionSystem/Assignment2_Image/512test1star.bmp'
image1=cv2.imread(path1)
path2 = '/content/gdrive/MyDrive/차량비전시스템/VisionSystem/Assignment2_Image/abcdef.bmp'
image2=cv2.imread(path2)

# 이미지 크기 조정 - 안하면 이미지의 크기가 다르기 때문에 AND, OR, XOR 연산 불가, 오류 발생
image1 = cv2.resize(image1, (image2.shape[1], image2.shape[0]))

# ① AND, OR, XOR 연산 수행 후 각 결과 이미지 출력
#  제출물 - 읽어온 이미지, 각 논리연산 결과 이미지

print("① AND, OR, XOR 연산 수행 후 각 결과 이미지 출력")

print(" 제출물")

print(" 읽어온 이미지 :")
cv2_imshow(image1)
cv2_imshow(image2)

print(" 각 논리연산 결과 이미지 :")
result_and = cv2.bitwise_and(image1, image2)
result_or = cv2.bitwise_or(image1, image2)
result_xor = cv2.bitwise_xor(image1, image2)

cv2_imshow(result_and)
cv2_imshow(result_or)
cv2_imshow(result_xor)
```

```python
## 3. RGB 이미지 (최소 2개)를 읽은 후 Gray scale로 변환
path3 = '/content/gdrive/MyDrive/차량비전시스템/VisionSystem/Assignment2_Image/cat.bmp'
image3=cv2.imread(path3)
path4 = '/content/gdrive/MyDrive/차량비전시스템/VisionSystem/Assignment2_Image/mandrill.jpg'
image4=cv2.imread(path4)

# ① 전역임계값 적용후 영상 내에서의 객체 분할
#  제출물 - 읽어온 RGB 이미지 및 Gray scale로 변환한 이미지, 전역임계값 적용 후 객체 분할 결과
print("① 전역임계값 적용후 영상 내에서의 객체 분할")

print(" 제출물")

print(" 읽어온 RGB 이미지 :")
cv2_imshow(image3)
cv2_imshow(image4)

print(" Gray scale로 변환한 이미지 :")
gray_image3 = cv2.cvtColor(image3, cv2.COLOR_BGR2GRAY)
gray_image4 = cv2.cvtColor(image4, cv2.COLOR_BGR2GRAY)

cv2_imshow(gray_image3)
cv2_imshow(gray_image4)

print(" 전역임계값 적용 후 객체 분할 결과 :")
_, thresh3 = cv2.threshold(gray_image3, 127, 255, cv2.THRESH_BINARY)
_, thresh4 = cv2.threshold(gray_image4, 127, 255, cv2.THRESH_BINARY)

cv2_imshow(thresh3)
cv2_imshow(thresh4)
```

```python
# ② 적응적 임계값 적용후 영상 내에서의 객체 분할
#  제출물 - 적응적 임계값 적용 후 객체 분할 결과
print("② 적응적 임계값 적용후 영상 내에서의 객체 분할")

print(" 제출물")

print(" 적응적 임계값 적용 후 객체 분할 결과 :")
adaptive_thresh3 = cv2.adaptiveThreshold(gray_image3, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11, 2)
adaptive_thresh4 = cv2.adaptiveThreshold(gray_image4, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11, 2)

cv2_imshow(adaptive_thresh3)
cv2_imshow(adaptive_thresh4)
```

```python
## 4. RGB 이미지(최소 2개)를 읽은 후 Gray scale로 변환
path5 = '/content/gdrive/MyDrive/차량비전시스템/VisionSystem/Assignment2_Image/medicine.bmp'
image5=cv2.imread(path5)
path6 = '/content/gdrive/MyDrive/차량비전시스템/VisionSystem/Assignment2_Image/Copy_of_test2.png'
image6=cv2.imread(path6)

gray_image5 = cv2.cvtColor(image5, cv2.COLOR_BGR2GRAY)
gray_image6 = cv2.cvtColor(image6, cv2.COLOR_BGR2GRAY)

# ① 히스토그램 스트레칭 수행
#  제출물 - 히스토그램 스트레칭 적용 전/후 이미지
print("① 히스토그램 스트레칭 수행")

print(" 제출물")

print(" 히스토그램 스트레칭 적용 전/후 이미지 :")
min_intensity5 = np.min(gray_image5)
max_intensity5 = np.max(gray_image5)
stretched_image5 = (255 / (max_intensity5 - min_intensity5) * (gray_image5 - min_intensity5)).astype(np.uint8)

min_intensity6 = np.min(gray_image6)
max_intensity6 = np.max(gray_image6)
stretched_image6 = (255 / (max_intensity6 - min_intensity6) * (gray_image6 - min_intensity6)).astype(np.uint8)

cv2_imshow(gray_image5)
cv2_imshow(stretched_image5)
cv2_imshow(gray_image6)
cv2_imshow(stretched_image6)
```

```python
# ② 히스토그램 평활화 수행
#  제출물 - 히스토그램 평활화 적용 전/후 이미지

print("② 히스토그램 평활화 수행")

print(" 제출물")

print(" 히스토그램 평활화 적용 전/후 이미지 :")
equalized_image5 = cv2.equalizeHist(gray_image5)
equalized_image6 = cv2.equalizeHist(gray_image6)

cv2_imshow(gray_image5)
cv2_imshow(equalized_image5)
cv2_imshow(gray_image6)
cv2_imshow(equalized_image6)
```
